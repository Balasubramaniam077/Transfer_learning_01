**Importing all requirement files for Transfer Learning**

```python
import time

import torch
import torch.optim as optim
import torch.nn as nn
import torchvision.datasets as dset
import torchvision.transforms as transforms
import torch
from torch.autograd import Variable
import torchvision
from torchvision import models
```

**Preprocessing the transforms to perform the Data**

```python
#In my code i wish to set an batch size into 32 if you need to change to change to 16,3,64 

batch_size = 32
lr = 0.001
```

```python
def trainDataset():
    
    #I given the Dataset training path for training Folder 
    dataset = dset.ImageFolder(root="../input/100-bird-species/train",
                            transform=transforms.Compose([
                                transforms.RandomRotation(30),
                                transforms.RandomHorizontalFlip(),
                                transforms.RandomResizedCrop(224, scale=(0.96, 1.0), ratio=(0.95, 1.05)),
                                transforms.ToTensor(),
                                transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
                            ]))
    dataloader = torch.utils.data.DataLoader(dataset,
                                            batch_size=batch_size,
                                            shuffle=True,
                                            num_workers=4)
    return dataloader                               

def testDataset():
    #I given the dataset folder path for Testing Folder 
    dataset = dset.ImageFolder(root="../input/100-bird-species/test",
                            transform=transforms.Compose([
                                transforms.Resize([224, 224]),
                                transforms.ToTensor(),
                                transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
                            ]))
    dataloader = torch.utils.data.DataLoader(dataset,
                                            batch_size=batch_size,
                                            shuffle=True,
                                            num_workers=4)
    return dataloader
```

```python
#checking the cuda is available or not 

device = torch.device("cuda")
```

```python
#dataset input way i already Given the dataset path in preprocessing script 

train_loader = trainDataset()
test_loader = testDataset()
```

```python
#define the Resnet model for training the dataset 

model = models.resnet50().cuda()

#Loss function is CrossEntropy if you need to checking with various loss function also CrossEntropy is my choices
criterion = nn.CrossEntropyLoss().cuda()


#optimizer is an Adam if you want yo seena an result with various optimizer is alsio possible with my code
optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-5)
```

```python
def train(epoch):
    model.train()
    for idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)
        optimizer.zero_grad()
        output = model(data)
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()

        if idx % 10 == 0:
            print("epoch: ", epoch, "  process: ", int((idx / len(train_loader)) * 100),
                "%  Loss: ", loss.data.item())
    
    torch.save(model.state_dict(), './Resnet50'+str(epoch)+'.pt')

def test():
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += criterion(output, target).data.item()
            _, predicted = torch.max(output, 1)
            correct += (predicted == target).sum().item()
    
    test_loss /= len(test_loader.dataset)
    print("Average Loss: ", test_loss, "  Accuracy: ", correct, " / ",
    len(test_loader.dataset), "  ", int(correct / len(test_loader.dataset) * 100), "%")


 #if you want to increase or decrease the Epoches please the epoches value in For Loop 
    
if __name__ == "__main__":
    for epoch in range(3):
        start = time.time()
        train(epoch)
        end = time.time()
        print("It takes ", end - start, " seconds")
        test()
```
